- monitor: quantity to be monitored.

- factor: factor by which the learning rate will be reduced. new_lr = lr * factor.

- patience: number of epochs with no improvement after which learning rate will be reduced.

- verbose: int. 0: quiet, 1: update messages.


- min_lr: lower bound on the learning rate.


- __ReduceLearnigOnPlateau__ : we will decrease the learning rate if there is no improvment on the val_score (at Plateau).


 __EarlyStopping__ :
- if there is no improvement in val_accuracy then we will stop training process.
- or if val_loss is not decrease then we will stop at particular epoch.
